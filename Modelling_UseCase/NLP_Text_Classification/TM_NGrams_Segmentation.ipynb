## Topic Modelling

#### Create Topic Modelling (Key topics that review) and Segemention (Create diffrent cluster for similar review)

###  Data Preparation

```python
#imports
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data into a DataFrame
data = pd.read_csv("C:/Users/om\Downloads/Text mining & NLP files/Case Study 4 - Unsupervised Model on Text Data (Yelp)/reviews.csv")

#make a copy of columns I need from raw data
df1 = data.iloc[:, [3,4,5,6,8]]
```

```python
df1 = df1.rename(columns={"stars":"Score","text":"Text"})
df1.head()
```

```python
df1['Bad'] = np.where(df1.Score<=3,1,0) #if review is less then 3 bad else good
```

```python
# convert text to lowercase
df1.loc[:, 'Text'] = df1['Text'].str.lower()
```

```python
# get the tf-idf table 

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df = 0.05, max_df=0.95,
                             ngram_range=(1, 1), 
                             stop_words='english')
vectorizer.fit(df1['Text'])
```

```python
# Create a training data
X_train = vectorizer.transform(df1['Text'])
```

### Topic Modeling

```python
# ?decomposition.LatentDirichletAllocation
```

```python
# train a LDA Model

from sklearn import decomposition

lda_model = decomposition.LatentDirichletAllocation(n_components=5, learning_method='batch', max_iter=100) 
#we can create 5, 10 ...component
#learning_method='online' and batch. In batch certain batch go, bydefault 120 batch & analaysis but online is continous process
    #when we have large amout of data, online will be faster then batch.
X_topics = lda_model.fit_transform(X_train)
topic_word = lda_model.components_ #collection of terms 
vocab = vectorizer.get_feature_names() #all the terms used 
```

```python
np.array(vocab)
```

```python
topic_word #under each component important of each word, similar to PCA
```

```python
#Number of rows is equal to number of componets and column is equal to number of voca
```

```python
# view the topic models
n_top_words = 10 #Give the top 10 word from each componets 
topic_summaries = []
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-(n_top_words+1):-1] 
    topic_summaries.append(' '.join(topic_words))

topic_summaries

#enumerate automatically cal the range of word
#np.array(vocab)[np.argsort(topic_dist)] sorting the data with its importance, [:-(n_top_words+1):-1 
        #give the top 10 (sorting will be asc)--->[:,-10:-1]
#storing data in topic_summaries
```

```python
#These are not sentance, these are giving importantant word frequently occured 
#1 componet is more about food and service 
#2 componet is more about bar
#3 particular type of food (pizza)
#3 store , place
#4 food order
#changing the learning method like batch n online will also impcate the result
```

### Segmentation

```python
#Apply TfidfVectorizer to review text
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
# we create a kmeans model
model = KMeans(n_clusters=4, 
               init='k-means++', 
               max_iter=100, n_init=1,random_state=5)
```

```python
# apply the model on this prepared data
res = model.fit(X_train)
res
```

```python
# Getting the vocalbulary as well as cluster centers
# Firstly, set of words from the tf-idf itself
vocab = vectorizer.get_feature_names()
vocab = np.array(vocab)
```

```python
# Then, the cluster centers from the model fit that we stored on res
cluster_centers = np.array(res.cluster_centers_)
cluster_centers #we have created 4 cluster mean we have 4 centroid --> no. of rows is 4 and col are voca
```

```python
print(np.shape(res.cluster_centers_))
print(np.shape(res.cluster_centers_)[0])
```

```python
# get top 10 words from that cluster(4)
words=set()
for i in range(len(res.cluster_centers_)):
    words = set(vocab[sorted_vals[i][-10:]]) #top 10 word for each cluster 
    print(words)
```
