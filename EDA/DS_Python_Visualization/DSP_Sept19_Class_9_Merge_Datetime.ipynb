```python
visit = pd.read_excel('D:/SampleData/Visits.xlsx')
demog = pd.read_excel('D:/SampleData/Demog.xlsx')
```

```python
visit
```

```python
demog
```

```python
# Get the details of registered patients and date of their visits?
pd.merge(left = demog, right=visit, on = 'PATIENT_ID', how = 'left')
```

```python
# Get the list of patients who have not yet registered with us?
mer_data  =pd.merge(left = demog, right=visit, on = 'PATIENT_ID', how = 'right', indicator=True)
```

```python
mer_data[mer_data._merge == 'right_only']
```

```python
# Get the list of registered patients who have never make a visit?
pd.merge(left = demog, right=visit, on = 'PATIENT_ID', how = 'left', indicator=True)
```

```python
visit.rename(columns = {'DOV': 'Date'}, inplace=True)
```

```python
demog.rename(columns = {'DOB': 'Date'}, inplace=True)
```

```python
pd.merge(left = demog, right=visit, on = 'PATIENT_ID', how = 'outer', suffixes=('_demog', '_visit'))
```
