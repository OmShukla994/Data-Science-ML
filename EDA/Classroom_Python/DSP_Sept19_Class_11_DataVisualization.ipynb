```python
def fn_distplot(x):
    plt.figure(figsize=(5, 3))
    sns.distplot(x.dropna())
    plt.show()
```

```python
def fn_countplot(x):
    plt.figure(figsize=(5, 3))
    sns.countplot(x.dropna())
    plt.show()
```

```python
stores.select_dtypes(['object']).apply(fn_countplot)
```

```python
stores.select_dtypes(['float64', 'int64']).apply(fn_distplot)
```
