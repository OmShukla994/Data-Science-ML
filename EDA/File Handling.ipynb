```python
## We can also split lines using file handling in Python. 
#This splits the variable when space is encountered.
with open("python.txt", "r") as file: 
    data = file.readlines() 
    for line in data: 
        word = line.split() 
        print (word )
```

### Ques: Program to find largest word in a file

```python
#Program to print largest word in file:
with open("ashi.txt") as file:
    words=file.read().split()
    print(words)
    max_len=len(max(words,key=len))
    for word in words:
        if len(word)==max_len:
            print("The longest word in file = %s and its length = %d :"%(word,max_len))

```

### Ques : A Python program to get the file size of a plain file

```python
def file_size(fname):
        import os
        statinfo = os.stat(fname)
        return statinfo.st_size

print("File size in bytes of a plain file= ",file_size("python.txt"))
```
