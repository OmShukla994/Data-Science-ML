# What is Machine Learning?

Machine learning is an application that provides Computers the ability to automatically learn and improve from experience without being explicitly programmed.
- ML is a technique to train machines on a particular algorithm.
- Major part of ML includes Mathematics with Statistics & Probability.
- Python is one of the most strongest language for ML ,with many inbuilt libariries.


 



## How to approach?
- Find out the problem e.g : to predict stock price , spam or not spam , pass or fail etc.
- Gather related (Data)information
- Data treatment(EDA)
- Training Data with ML algorithms(Regression , Logistic regression , KNN ,decision tree etc.)
- Present Results
- Making Predictions
- Buisness Implemetation

## Important Libraries 
- Numpy for Numerical Analysis 
- Pandas for Data Preprocessing & Visualization
- Matplotlib , Seaborn & Bokeh for Data Visualization
- Scikit-learn for  ML algorithms
- Tensor Flow , Theano & Keras for Deep Learning
- NLTK for Natural language preprocessing 
- StatsModel for the estimation of many different statistical models, as well as for conducting statistical tests, and statistical data exploration.
